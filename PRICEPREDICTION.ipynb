```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import matplotlib
```

```python
df1 = pd.read_csv("/content/bengaluru_house_prices.csv")
df1.head()
```

```python
df1.groupby('area_type')['area_type'].agg('count')
```

```python
df2 = df1.drop(['area_type','society','balcony','availability'],axis = 'columns')
df2
```

Data Cleaning Starts here....................

```python
df2.isnull().sum()
```

```python
df3 = df2.dropna()
df3.isnull().sum()
```

```python
df3['size'].unique()
```

```python
df3['bhk'] = df3["size"].apply(lambda x : int(x.split(" ")[0]))
df3
```

```python
df3['bhk'].unique()
```

```python
df3[df3['bhk']>20]
```

```python
def convert(x):
  tokens = x.split('-')
  if len(tokens) == 2:
    return (float(tokens[0])+float(tokens[1]))/2
  try:
    return float(x)
  except :
    return None
```

```python
df4  =df3.copy()
df4['total_sqft'] = df4['total_sqft'].apply(convert)
df4.head()
```

```python
 df5 = df4.copy()
 df5['price_per_sqft'] = df5['price']*100000/df5['total_sqft']
 df5.head()
```

```python
 len(df5.location.unique())
```

```python
location_stats = df5.groupby('location')['location'].agg('count').sort_values()
location_stats
```

```python
location_less_10 = location_stats[location_stats<=10]
location_less_10
```

```python
df5.location = df5.location.apply(lambda x: 'other location' if x in location_less_10 else x)
len(df5.location.unique())
```

```python
df5[df5.total_sqft/df5.bhk<300].head()
```

```python
df6 = df5[~(df5.total_sqft/df5.bhk<300)]
```
